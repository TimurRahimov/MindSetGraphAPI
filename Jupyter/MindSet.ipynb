```python
import pandas as pd
from nebula_driver import NebulaDriver
```

```python
def preparing_data(csv_path='data_test.csv') -> pd.DataFrame:
    data = pd.read_csv(csv_path, sep=';')    
    return data
```

**Инициализация данных и переменных**

1) Загрузка данных из CSV в DataFrame

```python
data = preparing_data('data_test.csv')
```

2. Создание в БД необходимых пространств, сущностей и связей

```python
NebulaDriver.create_objects_in_database()
```

3. Внесение полученных данных в БД

```python
NebulaDriver.insert_data(data)
```

4. Построение графового представления БД в .graphml и сохранение его в файл

```python
graphml = NebulaDriver.get_all_in_graphml()
with open("for_gephi.graphml", "w") as f:
    f.write(graphml)
```
